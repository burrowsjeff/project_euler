## Highly divisible triangular number

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

* 1: 1
* 3: 1,3
* 6: 1,2,3,6
* 10: 1,2,5,10
* 15: 1,3,5,15
* 21: 1,3,7,21
* 28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

## Final Solution

Now that we have it working for the stated problem, lets expand it to the desired outcome.

```python
t = 1
c = 1
factor_list = [1,3]
x = 0

while len(factor_list) < 500:
    for i in range(c,1,-1):
        t = t + i
    factor_list = []
    for g in range(1,t):
        if t % g == 0:
            factor_list.append(g)
    factor_list.append(t)
    x = t
    c += 1
    t = 1
    
print('Triangle number is',x,', the number of factors is',len(factor_list),', and the factors are',factor_list)
```
